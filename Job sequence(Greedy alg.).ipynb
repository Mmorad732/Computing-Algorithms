```python
def sort(array,p,w):
  for x in range(0,len(array)-1):
    min = array[x]
    for y in range(x+1,len(array)):
      if min < array[y]:
        min = array[y]
        pmin = p[y]
        wmin = w[y]
    
        flag = y
        
    if min != array[x]:  
      array[flag]=array[x]
      array[x] = min

      p[flag] = p[x]
      p[x] = pmin

      w[flag] = w[x]
      w[x] = wmin


     
  
  return array , p , w 

def min(n1,n2):
  if n1>n2:
    return n2
  else:
    return n1  



def jobsequence(jobs , p , dl):
  #first sort the jobs according to the profit
  p,jobs,dl = sort(p,jobs,dl)
  print("Profit" , p)
  print("Jobs" , jobs)
  print("deadlines", dl)
  #Get the max deadline
  maxdeadline = 0 
  for x in range(len(dl)):
    if maxdeadline < dl[x]:
      maxdeadline = dl[x]
  #Create the time slot array according to the max deadline
  ts = [0 for x in range(maxdeadline+1)]
  
  #Put the jobs into the time slot array
  for i in range(len(p)):
    #Get min between max deadline and current deadline to 
    # decide where to put the current job 
    k = min(maxdeadline , dl[i])
    #Check if there is empty time slots that fits the current deadline
    while k >= 1:
      if ts[k] == 0 and k >= dl[i]:
        ts[k] = jobs[i]
        break
      k = k-1
  #Crop the extra part from array    
  ts = ts[1:maxdeadline+1]    
  print("time slot\t",ts)    





#Assign jobs to time slot according to gain the most profit

jobs= ['j1','j2','j3','j4','j5','j6','j7']
profit = [35,30,25,20,15,12,5]
deadline = [3,4,4,2,3,1,2]
jobsequence(jobs,profit,deadline)
  

  
```
