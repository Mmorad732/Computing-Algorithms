```python
array=[6,10,1,5,4,6]

for x in range(0,len(array)-1):
  min = array[x]
    
  for y in range(x+1,len(array)):
    if min > array[y]:
      min = array[y]
      flag = y
      
  if min != array[x]:  
    array[flag] = array[x]
    array[x] = min
  print("Path ",x+1," array = ",array)  

 
   
  


    
```
