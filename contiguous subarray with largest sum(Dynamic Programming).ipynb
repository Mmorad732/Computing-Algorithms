```python
def dp(a):
  s = [0 for x in range(len(a))]
  s[0] = a[0]
  for i in range(1,len(s)):
    if a[i]+s[i-1] > a[i]:
      s[i] = a[i]+s[i-1]
    else:
      s[i] = a[i]
  max = s[0]
  for x in range(len(s)):
    if max<s[x]:
      max = s[x]

  return max

a = [-2,1,-3,4,-1,2,1,-5,4]
print(dp(a))
```
