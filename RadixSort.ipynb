```python
def countSort(arr,pos):
  count=[]
  newArray = []
  for i in range(0,10):
    count.append(0)
  for i in arr:
    newArray.append(0)  

  for i in arr:
    index = int((i/pos)%10)
    count[index]=count[index]+1

  

  for i in range(1,len(count)):
    count[i]=count[i]+count[i-1]
 

  for i in range(len(arr)-1,-1,-1):
    index = int((arr[i]/pos)%10)
    count[index] = count[index]-1
    newArray[count[index]] = arr[i]

  return newArray 









arr=[122,423,311,424,542,645,266,343,432,325,239]
max=0
for i in arr:
  if max<i:
    max=i
print("max=",max)

pos = 1
while int(max/pos) > 0:
  arr=countSort(arr,pos)
  print(arr)
  pos=pos*10
  

```
