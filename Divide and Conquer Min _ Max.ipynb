```python
array = [2,5,7,20,8,40,50,1,30]
max = array[0]
min = array[0]


def minmax(a,low,high):
  global max , min 
  mid = int((low+high)/2)
  if low == mid:
    if a[low] > a[high]:
      temp = a[low]
      a[low] = a[high]
      a[high] = temp
    if min > a[low]:
      min = a[low]
    if max < a[high]:
      max = a[high]

  else:
    minmax(a,low , mid)
    minmax(a,mid+1,high)      












def main():
  global array ,max ,min
  minmax(array,0,len(array)-1)
  print('min =',min)
  print('max =',max)

if __name__ == "__main__":
  main()
```
