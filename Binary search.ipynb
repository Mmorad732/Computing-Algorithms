```python
import math as num

def binarySearch(n,array):
  flag = False
  i=0
  j=len(array)-1
  mid = int((i+j)/2)
  while i<=j:
    if array[mid]==n:
      flag = True
      break
    elif n < array[mid]:
      j=mid-1
    else:
      i=mid+1
    mid = int((i+j)/2)  
  if flag == True:
    return "True"
  else:
    return "false"          

array = [1,2,3,5,8,10,15,20,26]
print(binarySearch(15,array))
```
