```python
array=[15,16,6,8,5]

for y in range(0,len(array)):
  flag = False
  for x in range(0,len(array)-1-y):
    if (array[x]>array[x+1]):
      temp = array[x+1]
      array[x+1]=array[x]
      array[x]=temp
      flag=True
    print("Pass ",y+1," ", array)  
  if flag == False:
    break    
     
    


```
