```python
final_array = []

def partition(array,low,high):
  
  pivot= array[low]
  i = low 
  j = high
  while i < j:
     if array[i] <= pivot:
       i = i + 1
       

     if array[j] > pivot:
       j = j - 1
         
     
     if i < j:
       temp = array[i]
       array[i] = array[j]
       array[j] = temp
  temp = array[low]
  array[low] = array[j]
  array[j] = temp
  
  return j , array

def quicksort(array,low,high):
  arr = []
  if low < high:
    piv ,array = partition(array,low,high)
    quicksort(array,low,piv-1)
    quicksort(array,piv+1,high)
  if low > high:
    global final_array
    final_array = array
    



def main():
  arr= [12,3,3,2,4,5,21,76,32,4,7,8,5]
  arr = quicksort(arr , 0 , len(arr)-1)
  print(final_array)

if __name__ == "__main__":
  main()  
```
