```python
arr=[9,12,3,4,5,2,2,3,2,12,9,8,3,4,0,9,11]
min=0
max=0
for i in arr:
  if min > i:
    min = i
for i in arr:
  if max<i:
    max = i 


newArray=[]  

for i in arr:
  newArray.append(0)

count=[]
for i in range(max+1):
  count.append(0)

for i in arr:
  count[i]=count[i]+1


for i in range(1,len(count)):
  count[i]=count[i]+count[i-1]


for i in range(len(arr)-1,-1,-1):
  count[arr[i]] = count[arr[i]]-1
  newArray[count[arr[i]]] = arr[i]
  

print(newArray)
```
