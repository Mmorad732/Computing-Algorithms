```python
def sort(array,p,w ,i):
  for x in range(0,len(array)-1):
    min = array[x]
    for y in range(x+1,len(array)):
      if min > array[y]:
        min = array[y]
        pmin = p[y]
        wmin = w[y]
        imin = i[y]
        flag = y
        
    if min != array[x]:  
      array[flag]=array[x]
      array[x] = min

      p[flag] = p[x]
      p[x] = pmin

      w[flag] = w[x]
      w[x] = wmin


      i[flag] = i[x]
      i[x] = imin
    
  
  return array , p , w , i




def fractionalKnapsack(items,profit,w,capacity):
  
  ppw = [0 for x in range(len(items))]
  tp = 0
  tw = 0
  #Create profit per weight array (profit/weight)
  for i in range(len(ppw)):
    ppw[i] = float('%.2f'%(profit[i] / w[i]))
  #Sort all the arrays according to the  profit per weight
  ppw , profit , w , items =sort(ppw , profit , w , items) 
  
  #Assign the items to the fractional knapsack to gain max profit
  for x in range(len(ppw)-1 , 0,-1 ):
    #Check if there is capacity for the item then add item profit 
    # to the total profit
    if capacity - w[x] >= 0:
      capacity = capacity - w[x]
      tp = tp + p[x]
      tw = tw + w[x]
    else:
      fraction = capacity / w[x]  
      tp = float('%.2f'%(tp + (fraction*profit[x])))
      tw = float('%.2f'%(tw + (fraction*w[x])))


  print("items : ",items)
  print("Profit : ", profit)
  print("Weight : ", w)
  print("Profit per weight : ",ppw)
  print("Total profit : " , tp)
  print("Total weight : " , tw)



items = [1,2,3,4,5,6,7]
p = [10,5,15,7,6,18,3]
w = [2,3,5,7,1,4,1]
c = 15
fractionalKnapsack(items,p,w,c)
```
